```python
def visualize_cluster_folium(df):
    #35.8715163, 128.5959431
    df_drop = df.drop_duplicates()
    g_map = folium.Map(location = [35.8715163, 128.5959431], zoom_start=12, tiles = 'Stamen Terrain')
    color_list = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray']
    for index, row in df_drop.iterrows():
#         print(df_drop['ftr1'][i], ', ', df_drop['ftr2'][i])
#         print(color_list[df_drop['dbscan_cluster'][i]])
        if row['dbscan_cluster']>-1:
            lat = row['ftr1']
            lng = row['ftr2']
            color = color_list[int(row['dbscan_cluster'])%len(color_list)]
            folium.Marker([lat, lng], icon=folium.Icon(color=color)).add_to(g_map)
    return g_map
```

# **1.대구 명소 외국인 OPTICS(2018)**

```python
df=pd.read_csv('/content/drive/Shareddrives/[논문] 내 외국인관광객 유동인구 비교/분석/raw data/크롤링 결과/명소/3.연도별 구분/외국인/대구명소_2018_외국인.csv', encoding='cp949')
geo = df[['lat', 'lng']]
# geo_mul = geo.mul(1000000)
quot = min_samples=geo.shape[0]//20
min_samples = quot if quot>0 else 1
clustering = OPTICS(min_samples=min_samples).fit(geo)
dbscan_labels = clustering.labels_

geo['dbscan_cluster'] = dbscan_labels
geo.columns = ['ftr1', 'ftr2', 'dbscan_cluster']
#파일로 저장
df['dbscan'] = geo['dbscan_cluster']
df.to_csv('/content/drive/Shareddrives/[논문] 내 외국인관광객 유동인구 비교/분석/분석/OPTICS/OPTICS결과/OPTICS_대구명소_2018_외국인.csv', index=False, encoding='cp949')
#시각화
g_map = visualize_cluster_folium(geo)
print(len(np.unique(dbscan_labels)))
g_map
```
