```python
df = pd.read_csv(file+'.csv', sep=',', encoding='cp949')
df.head()
```

```python
df_drop = df.dropna(subset=['address'])
```

```python
def getLatLng(address, driver):
    url = ('https://www.google.co.kr/maps/search/'+address+
               '/@35.8698753,128.5880298,15z/data=!3m1!4b1?hl=ko')
    driver.get(url)
    time.sleep(2.5)
    try:
        button = driver.find_element_by_xpath("//button[@jsaction='pane.placeActions.directions']")
        button.click()
        time.sleep(2)
        urlSplit = driver.current_url.split('!')
        latitude = urlSplit[-2][2:]
        longtitude = urlSplit[-1].split('?')[0][2:]
        return latitude, longtitude
    except:
        return -1, -1

def fillTripadvisor(df):
    driver = webdriver.Chrome()
    driver.implicitly_wait(5)
    lats = list()
    lngs = list()
    for address in df['address']:
        lat, lng = getLatLng(address, driver)
        lats.append(lat)
        lngs.append(lng)
    df['lat'] = lats
    df['lng'] = lngs
    return df
```
